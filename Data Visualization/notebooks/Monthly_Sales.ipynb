```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
```

```python
'''
Sales data
sales_data = {
    Month_name: ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Product_A': [64, 120, 130, 110, 150, 140, 160, 170, 180, 200, 190, 85],
    'Product_B': [69, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 94],
    'Product_C': [74, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150, 48]}
    '''
```

```python
# Define parameters
num_months = 12  # Number of months
num_products = 3  # Number of products
min_sales = 50  # Minimum sales value
max_sales = 200  # Maximum sales value

# Generate random sales data
sales_data = {
    'Month': [f'{i+1}' for i in range(num_months)]
}

for product_num in range(num_products):
    product_sales = np.random.randint(min_sales, max_sales+1, size=num_months)
    sales_data[f'Product_{chr(65+product_num)}'] = product_sales

# Convert dictionary to DataFrame

sales_df=pd.DataFrame(sales_data)

sales_df.index = np.arange(1, len(sales_df) + 1)

# Replace numeric months with their corresponding names
sales_df['Month'] = pd.to_datetime(sales_df['Month'],format='%m').dt.month_name()

# Display the generated dataset
print(sales_df)
```

```python
# Convert dictionary to DataFrame
df = pd.DataFrame(sales_data)
```

```python
# Line plot to visualize sales trends over months for each product
plt.figure(figsize=(10, 6))
sns.lineplot(data=df.drop('Month', axis=1), markers=True)
plt.title('Monthly Sales Trends for Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Product')

#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Analysis:

Analysis: In the start of January, the sales of **Product A** and **Product C** falls down whereas the sales of **Product B** rises **from 80 to 140**. Initially the sales of A are highest.
The sales of **Product B** rises during the month of August and is around 160 in the end of the year.
The sales of the **Product B** are at its highest during the month of **May and July** but it falls down lower during the month of **June** and lowest in **August**.
**Product A** is having Highst stakes in **November**

```python
# Box plot to identify distribution of sales for each product
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('Month', axis=1))
plt.title('Distribution of Monthly Sales for Products')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Heatmap to visualize correlation between products
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop('Month', axis=1).corr(), annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation between Products')
plt.tight_layout()
plt.show()
```
