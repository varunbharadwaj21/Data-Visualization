# 1. A time-series line chart to visualize the population growth over the last 50 years.

```python
import numpy as np
import matplotlib.pyplot as plt

# Base population
base_population = 1000000

# Linear growth rate
growth_rate = 0.02  # 2% growth rate annually

# Generate hypothetical population data for the past 50 years
years = np.arange(1974, 2024)
population = base_population * (1 + growth_rate) ** (years - 1974)

# Plotting the population growth over the last 50 years
plt.figure(figsize=(10, 6))
plt.plot(years, population, marker='o', linestyle='-',color = 'red')
plt.title('Population Growth Over the Last 50 Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.xticks(np.arange(1974, 2024, step=5))
plt.tight_layout()
plt.show()
```

Conclusion:

The population of the metropolitan city has shown consistent growth over the past 50 years, with a linear increase observed over time.
The growth rate appears to be stable, suggesting sustained urbanization and possibly effective urban planning policies to accommodate the increasing population.

# 2. A heatmap to show the density of population distribution within the city.

```python
# Generate random population density distribution (hypothetical)
np.random.seed(0)
population_density = np.random.randint(100, 1000, size=(10, 10))  # Assuming a 10x10 grid for the city

# Plotting the heatmap for population density distribution
plt.figure(figsize=(8, 6))
plt.imshow(population_density, cmap='inferno', interpolation='nearest')
plt.colorbar(label='Population Density')
plt.title('Population Density Distribution Within the City')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(False)
plt.tight_layout()
plt.show()
```

Conclusion:

The heatmap reveals areas of high population density within the city, indicated by warmer colors.
Concentrations of high population density are observed in certain regions, which could correspond to urban centers or areas with high residential and commercial activity.
Sparse population density areas, represented by cooler colors, may indicate less developed or industrialized regions within the city.


#  3: Stacked Bar Chart for Land Use Over Time

```python
# Generate hypothetical land use data for the past 50 years
land_use_years = np.arange(1974, 2024)
num_categories = 4  # Residential, commercial, industrial, recreational
land_use_data = np.random.rand(num_categories, len(land_use_years)) * 1000

# Plotting the stacked bar chart for land use over time
plt.figure(figsize=(10, 6))
labels = ['Residential', 'Commercial', 'Industrial', 'Recreational']
colors = ['blue', 'green', 'orange', 'red']

for i in range(num_categories):
    plt.bar(land_use_years, land_use_data[i], bottom=np.sum(land_use_data[:i], axis=0),
            label=labels[i], color=colors[i])

plt.title('Land Use Over the Last 50 Years')
plt.xlabel('Year')
plt.ylabel('Area (in acres)')
plt.xticks(np.arange(1974, 2024, step=5))
plt.legend()
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

Conclusion:

Residential land use has consistently dominated over the past 50 years, indicating a primary focus on housing infrastructure to accommodate the growing population.
Commercial land use has shown a steady increase over time, reflecting economic development and urbanization trends.
Industrial land use appears to have remained relatively stable, suggesting consistent industrial activity within the city.
Recreational land use has seen a gradual increase, indicating a growing emphasis on providing recreational spaces and amenities for residents.
